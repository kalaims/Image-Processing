```python
im_gray = Image.open(img).convert('L') # image opened and convert to gray
im = np.asarray(im_gray) #image covert to an ndarry
print(type(im))
```

```python
# Display image and gray image
fx, plots = plt.subplots(1,2,figsize = (20,10))

plots[0].set_title("Original Image")
plots[0].imshow(image)

plots[1].set_title("Gray Image")
plots[1].imshow(im_gray, cmap="gray")
```

```python
gr_img= Image.fromarray(im).save('gr_WV.png') # converting NumPy array into PIL image with JPEG/PNG ext. Important step at the end to save image
```

```python
from skimage.feature import corner_fast, corner_peaks, corner_subpix
#img= np.ascontinuousarray(image)
img_response = corner_fast(im, n=12, threshold=0.001)
#img_response1= corner_fast(im, n=9, threshold=0.15)
#print(img_response)
cpv =corner_peaks(img_response,min_distance=1, threshold_rel=0.1)
corners_subpix_val= corner_subpix(im,cpv, window_size=11,alpha=0.99)
fig,ax = plt.subplots()
ax.imshow(im, cmap= plt.cm.gray) #interpolation = 'nearest'
x= corners_subpix_val[:,1]
y= corners_subpix_val[:,0]
ax.plot(x,y,'ob') #'ob',markersize=10
#ax.axis('off')
plt.savefig(r"C:\Users\kalai\RubikCube_out.png",dpi=300)
plt.show()
```
