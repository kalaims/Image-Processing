## Reading and Writing Images using PIL/Pillow
#### If pillow is not installed already, use the folowing command in your terminal..
##### > pip install pillow

```python
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
import os
from os import listdir
import glob
```

#### To load an image file, use the open() function in the Image module

```python
Img = Image.open(r"C:\Users\kalai\RubikCube.jpg") #Returns an image object
```

#### Using the instance attributes to examine the file contents

```python
print(Img)
```

```python
print(Img.format) # Prints the Image file format
print(Img.size)   # prints tuple containing width and height (in pixels)
print(Img.mode)   # prints numbers and names of bands in the image
```

#### Convert to gray image and save as different format

```python
im_gray = Img.convert('L') 
im_gray.save('gray_flower','png')
```

#### To display image

```python
Img.show() #displays image on a image viewer
```

```python
# Display original and gray image
fx, plots = plt.subplots(1,2,figsize = (20,10))

plots[0].set_title("Original Image")
plots[0].imshow(Img)

plots[1].set_title("Gray Image")
plots[1].imshow(im_gray, cmap="gray")
```

#### Read multiple image files

#os module :It allows you to interact with the system, and also contains utilities to construct or analyze file paths.It doesnt provide the feature to search files using wildcards. So not powerful as glob.

#glob module: Finds all the pathnames matching a specified pattern.

```python
#Using os module
os.chdir( 'C:/Users/kalai/myimages/') #changes the current working directory to the given path
files = os.listdir( 'C:/Users/kalai/myimages/') #list containing the names of the images given by path

for idx, file in enumerate(files):
    img = Image.open(file)
    plt.subplot(2,5,idx+1)
    plt.title(file.split('.')[0])
    plt.axis('off')
    plt.imshow(img)
```

```python
#Using glob
path =r"C:/Users/kalai/myimages/*.*"
files= glob.glob(path)

for x, file in enumerate(files) :
    img1 = Image.open(file)
    plt.subplot(2,5,x+1)
    plt.axis('off')
    plt.imshow(img1)
    
    
```
